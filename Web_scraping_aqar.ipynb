```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
```

# North of Riyadh

```python
response = requests.get('https://sa.aqar.fm/%D9%81%D9%84%D9%84-%D9%84%D9%84%D8%A8%D9%8A%D8%B9/%D8%A7%D9%84%D8%B1%D9%8A%D8%A7%D8%B6/%D8%B4%D9%85%D8%A7%D9%84-%D8%A7%D9%84%D8%B1%D9%8A%D8%A7%D8%B6')
#print(response.status_code)

soup = BeautifulSoup(response.text, 'html.parser')
housesDetails = soup.findAll(class_="titleAndDetails")

location='north'
sleepRoom=[]
numOfBaths=[]
houseSizes=[]
prices=[]

for i in np.arange(1,int(51)):
    url='https://sa.aqar.fm/%D9%81%D9%84%D9%84-%D9%84%D9%84%D8%A8%D9%8A%D8%B9/%D8%A7%D9%84%D8%B1%D9%8A%D8%A7%D8%B6/%D8%B4%D9%85%D8%A7%D9%84-%D8%A7%D9%84%D8%B1%D9%8A%D8%A7%D8%B6'+str(i)
    #response = requests.get(url)
    #soup = BeautifulSoup(response.text, 'html.parser')
    
    
#-----------------------------------------------prices--------------------------------------------------------

    for house in housesDetails:
        if house.find('span', class_='price') is not None:
            price = house.find('span',class_='price').text
            prices.append(price)
        else:
            prices.append(None)
        

#--------------------------------------------sleepRoom-----------------------------------------------------------

        if house.find('span', class_='bed') is not None:
            beds = house.find('span',class_='bed').text
            sleepRoom.append(beds)
        else:
            sleepRoom.append(None)

#--------------------------------------------bathRoom-----------------------------------------------------------

        if house.find('span', class_='bath') is not None:
            baths = house.find('span',class_='bath').text
            numOfBaths.append(baths)
        else:
            numOfBaths.append(None)


#--------------------------------------------houseSizes-----------------------------------------------------------

        if house.find('span', class_='size') is not None:
            size = house.find('span',class_='size').text
            houseSizes.append(size)
        else:
            houseSizes.append(None)
            
```

```python
df1=pd.DataFrame({"Location":location,'SleepRoom':sleepRoom,'BathRoom':numOfBaths,'Size':houseSizes,'Prices':prices})
df1
```
